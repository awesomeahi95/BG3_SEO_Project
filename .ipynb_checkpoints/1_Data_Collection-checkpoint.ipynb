## Scraping Using Selenium

Selenium is configured to simulate a browser and scrape the following data for each URL:
- **Main Body Content**: The primary text content of the guide.
- **Meta Title**: The page title from the `<title>` tag.
- **Meta Description**: The page description from the `<meta name="description">` tag.
- **Meta Keywords**: The keywords from the `<meta name="keywords">` tag (if available).

This data will be stored in a structured format for subsequent processing and analysis.

```python
chrome_options = Options()
chrome_options.add_argument("--headless") 
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chromedriver_path = r"C:\Users\candy\Desktop\chromedriver-win64\chromedriver.exe"  
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

driver.implicitly_wait(30)
driver.set_page_load_timeout(240)

def scrape_url(url, question):
    try:
        driver.get(url)
        time.sleep(5)
        WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.TAG_NAME, "body")))
        content = driver.find_element(By.TAG_NAME, "body").text
        meta_title = driver.execute_script("return document.title;")
        
        # Handle cases where meta tags might not exist
        meta_description = driver.execute_script("""
            var meta = document.querySelector('meta[name="description"]');
            return meta ? meta.getAttribute('content') : null;
        """)
        meta_keywords = driver.execute_script("""
            var meta = document.querySelector('meta[name="keywords"]');
            return meta ? meta.getAttribute('content') : null;
        """)
        
        print(f"Successfully scraped: {url}")
        return {
            "Google_Question": question,
            "URL": url,
            "Content": content,
            "Meta_Title": meta_title,
            "Meta_Description": meta_description,
            "Meta_Keywords": meta_keywords
        }
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return {
            "Google_Question": question,
            "URL": url,
            "Content": None,
            "Meta_Title": None,
            "Meta_Description": None,
            "Meta_Keywords": None
        }

scraped_data = []
for entry in questions_and_urls:
    question = entry["question"]
    urls = entry["urls"]
    for url in urls:
        scraped_data.append(scrape_url(url, question))

scraped_dataframe = pd.DataFrame(scraped_data)

driver.quit()

scraped_dataframe.head()
```
