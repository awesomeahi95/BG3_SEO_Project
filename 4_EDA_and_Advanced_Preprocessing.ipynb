# Exploratory Data Analysis and Advanced Preprocessing
## <ins>Objective</ins>
- Perform exploratory data analysis (EDA) to understand patterns, distributions, and trends in the content and meta tags (`Content`, `Meta_Title`, and `Meta_Description`) of the scraped guides.
- Identify and remove additional domain-specific or irrelevant stopwords based on the analysis.
- Apply advanced preprocessing steps, such as lemmatization and vectorization, to prepare the data for SEO insights or potential machine learning tasks.
## <ins>Workflow</ins>
1. Load preprocessed data from the previous notebook.
2. Analyze word frequency distributions across all text fields (`Content`, `Meta_Title`, `Meta_Description`).
3. Identify and remove additional domain-specific stopwords based on the analysis.
4. Perform lemmatization on all text fields to standardize word forms.
5. Explore trends in meta tags (e.g., length distributions, keyword usage).
6. Save the fully processed data for downstream tasks.

## <ins>Imports</ins>

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter

# Configure visualizations
plt.style.use('ggplot')
```

## <ins>Load Data</ins>

```python
data_filepath = "data/initial_preprocessed_data.pkl"
preprocessed_dataframe = pd.read_pickle(data_filepath)
```

## Step 2: Analyze Word Frequency Distributions

Word frequency analysis helps identify commonly used terms across all text fields, highlighting patterns and potential noise. This step will be performed separately for `Content`, `Meta_Title`, and `Meta_Description`.

```python
# Define a function to calculate and plot word frequencies
def plot_word_frequencies(column, title):
    all_words = " ".join(preprocessed_dataframe[column].dropna().astype(str)).split()
    word_counts = Counter(all_words)
    most_common = word_counts.most_common(20)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(*zip(*most_common))
    plt.title(f"Top 20 Most Frequent Words in {title}")
    plt.ylabel("Frequency")
    plt.xlabel("Words")
    plt.xticks(rotation=45)
    plt.show()

# Analyze word frequencies for all relevant columns
plot_word_frequencies("Content", "Content")
plot_word_frequencies("Meta_Title", "Meta_Title")
plot_word_frequencies("Meta_Description", "Meta_Description")
```

## Step 3: Remove Additional Domain-Specific Stopwords

Based on the word frequency analysis, domain-specific stopwords (e.g., common terms unrelated to SEO analysis) will be identified and removed to reduce noise in the dataset.

```python
# Extend the stopword list with domain-specific terms
custom_stopwords = set(['baldur', 'gate', '3', 'guide', 'get', 'use', 'game', 'character'])
stop_words = set(stopwords.words("english"))
stop_words.update(custom_stopwords)

# Remove domain-specific stopwords from all relevant columns
def remove_stopwords(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
        return " ".join(filtered_tokens)
    return text

preprocessed_dataframe["Content"] = preprocessed_dataframe["Content"].apply(remove_stopwords)
preprocessed_dataframe["Meta_Title"] = preprocessed_dataframe["Meta_Title"].apply(remove_stopwords)
preprocessed_dataframe["Meta_Description"] = preprocessed_dataframe["Meta_Description"].apply(remove_stopwords)
```

## Step 4: Perform Lemmatization

Lemmatization reduces words to their base or root form, enabling more effective text analysis and reducing redundant terms.

```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function to lemmatize text
def lemmatize_text(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
        return " ".join(lemmatized_tokens)
    return text

# Apply lemmatization to all relevant columns
preprocessed_dataframe["Content"] = preprocessed_dataframe["Content"].apply(lemmatize_text)
preprocessed_dataframe["Meta_Title"] = preprocessed_dataframe["Meta_Title"].apply(lemmatize_text)
preprocessed_dataframe["Meta_Description"] = preprocessed_dataframe["Meta_Description"].apply(lemmatize_text)
```

## Step 5: Explore Meta Tag Trends

Meta tags play a crucial role in SEO. This step examines the length distributions and trends in `Meta_Title` and `Meta_Description`.

```python
# Calculate and visualize length distributions for meta tags
preprocessed_dataframe["Meta_Title_Length"] = preprocessed_dataframe["Meta_Title"].str.len()
preprocessed_dataframe["Meta_Description_Length"] = preprocessed_dataframe["Meta_Description"].str.len()

# Plot length distributions
plt.figure(figsize=(10, 6))
plt.hist(preprocessed_dataframe["Meta_Title_Length"].dropna(), bins=20, alpha=0.7, label="Meta_Title_Length")
plt.hist(preprocessed_dataframe["Meta_Description_Length"].dropna(), bins=20, alpha=0.7, label="Meta_Description_Length")
plt.title("Length Distributions of Meta Tags")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Step 6: Save Fully Processed Data

The fully processed data, including cleaned and lemmatized text fields and meta tag trends, will be saved for further analysis or modeling tasks.

```python
# Save the fully processed data
preprocessed_dataframe.to_pickle("data/fully_processed_data.pkl")
preprocessed_dataframe.to_csv("data/fully_processed_data.csv", index=False)

print("Saved the fully processed data for downstream tasks.")
```
